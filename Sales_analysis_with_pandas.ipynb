<h4> Exploratory data analysis

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#getting month year from the dateset
df['month_year']= df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
df.groupby("month_year").sum()
```

```python
df.groupby("month_year").sum()['sales']
```

```python
df_trend=df.groupby("month_year").sum()['sales'].reset_index()
```

```python
df_trend
```

```python

plt.plot(df_trend['month_year'],df_trend['sales'])
```

```python
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'],df_trend['sales'])
plt.xticks(rotation='vertical', size=8)
plt.show()
```

<h3> which are the top 10 products by sales?

```python
df.head()
```

```python
df['product_name']
```

```python
#grouping product name column
df.groupby('product_name').sum()['sales']
```

```python
prod_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
prod_sales
```

```python
prod_sales=prod_sales.sort_values('sales',ascending=False) 
```

prod_sales

```python
# top 10 by products by sales
prod_sales[:10]
```

<h3> Which are the most selling products

```python
#Grouping product name
most_sell_prod=pd.DataFrame(df.groupby("product_name").sum()['quantity'])
```

```python
most_sell_prod.sort_values('quantity',ascending=False)
```

```python
most_sell_prod[:10]
```
